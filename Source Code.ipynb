# Bilateral filter implementation sequential and parallel (On CPU) in C++

```python
%%writefile bilateral.cpp

#include <iostream>
#include <stdio.h>
#include <opencv2/opencv.hpp>
#include <thread>
#include <chrono>

using namespace cv;
using namespace std;

float distance(int x, int y, int i, int j) {
    return float(sqrt(pow(x - i, 2) + pow(y - j, 2)));
}

float gaussian(float x, double sigma) {
    //return exp(-(pow(x, 2))/(2 * pow(sigma, 2))) / (2 * CV_PI * pow(sigma, 2));
    return expf(-0.5f * (x*x) / (sigma* sigma));

}
void applyBilateralFilter(Mat source, Mat filteredImage, int x, int y, int diameter, double sigmaI, double sigmaS) {
    double iFiltered = 0;
    double wP = 0;
    int neighbor_x = 0;
    int neighbor_y = 0;
    int half = diameter / 2;

    for(int i = 0; i < diameter; i++) {
        for(int j = 0; j < diameter; j++) {
            neighbor_x = x - (half - i);
            neighbor_y = y - (half - j);
            double gi = gaussian(source.at<uchar>(neighbor_x, neighbor_y) - source.at<uchar>(x, y), sigmaI);
            double gs = gaussian(distance(x, y, neighbor_x, neighbor_y), sigmaS);
            double w = gi * gs;
            iFiltered = iFiltered + source.at<uchar>(neighbor_x, neighbor_y) * w;
            wP = wP + w;
        }
    }
    iFiltered = iFiltered / wP;
    filteredImage.at<double>(x, y) = (unsigned char)iFiltered;


}


Mat bilateralFilterSeq(Mat source, int diameter, double sigmaI, double sigmaS) {
    Mat filteredImage = Mat::zeros(source.rows,source.cols,CV_64F);
    int radius=0;

    int width = source.cols;
    int height = source.rows;

    for(int i = radius; i < height - radius; i++) {
        for(int j = radius; j < width - radius; j++) {
            applyBilateralFilter(source, filteredImage, i, j, diameter, sigmaI, sigmaS);
        }
    }
    return filteredImage;
}


Mat bilateralFilterParallel(Mat source, int diameter, double sigmaI, double sigmaS,int numThreads) {
    Mat filteredImage = Mat::zeros(source.rows, source.cols, CV_64F);
    int radius = 0;

    int width = source.cols;
    int height = source.rows;

    //int numThreads = thread::hardware_concurrency();
    
    vector<thread> threads;

    int rowsPerThread = height / numThreads;

    int start = radius;
    int end = start + rowsPerThread;

    for (int t = 0; t < numThreads; t++) {
        if (t == numThreads - 1) {
            end = height - radius;
        }

        threads.push_back(thread([&, start, end]() {
            for (int i = start; i < end; i++) {
                for (int j = radius; j < width - radius; j++) {
                    applyBilateralFilter(source, filteredImage, i, j, diameter, sigmaI, sigmaS);
                }
            }
        }));

        start = end;
        end = start + rowsPerThread;
    }

    for (auto &t : threads) {
        t.join();
    }

    return filteredImage;
}


int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter= atoi(argv[2]);
    double sigma_space= atoi(argv[3]);
    double sigma_intensity= atoi(argv[4]);
    int numThreads = 10;
    

    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageSeq = bilateralFilterSeq(src, diameter, sigma_intensity,sigma_space);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();


    start = chrono::high_resolution_clock::now();
    Mat filteredImageOpenCV;
    bilateralFilter(src, filteredImageOpenCV, diameter,sigma_intensity,sigma_space);
    end = chrono::high_resolution_clock::now();
    auto durationOpenCV = chrono::duration_cast<chrono::milliseconds>(end - start).count();

    start = chrono::high_resolution_clock::now();
    Mat filteredImageParallel = bilateralFilterParallel(src, diameter, sigma_intensity,sigma_space,numThreads);
    end = chrono::high_resolution_clock::now();
    auto durationParallel = chrono::duration_cast<chrono::milliseconds>(end - start).count();

    cout<<"Sequential: "<<durationSeq<<"ms \n";
    cout<<"Parallel: "<<durationParallel<<"ms"<<" using "<<numThreads<<" threads\n";

    imwrite("output/original_image_grayscale.png", src);
    imwrite("output/Bilateral_OpenCV.png", filteredImageOpenCV);
    imwrite("output/Bilateral_Seq.png", filteredImageSeq);
    imwrite("output/Bilateral_Parallel.png", filteredImageParallel);

    return 0;
}
```

# Bilateral filter implementation (On GPU) in C++

## Naive

```python
%%writefile bilateral_cuda.cu
#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <opencv2/opencv.hpp>

using namespace std;
using namespace cv;

// Function to calculate the gaussian weight based on intensity difference
__device__ float gaussianIntensity(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__device__ float gaussianSpace(float spatial_distance, float sigma_space) {
    return expf(-0.5f * (spatial_distance * spatial_distance) / (sigma_space * sigma_space));
}

__global__ void bilateralFilterKernel(const unsigned char* src, unsigned char* dst, int rows, int cols, int diameter, float sigma_space, float sigma_intensity) {
    int x = blockIdx.x * blockDim.x + threadIdx.x;
    int y = blockIdx.y * blockDim.y + threadIdx.y;

    if (x >= cols || y >= rows) {
        return;
    }

    float intensity_sum = 0;
    float weight_sum = 0;

    for (int j = -diameter; j <= diameter; j++) {
        for (int i = -diameter; i <= diameter; i++) {
            int neighbor_x = x + i;
            int neighbor_y = y + j;

            if (neighbor_x < 0 || neighbor_y < 0 || neighbor_x >= cols || neighbor_y >= rows) {
                continue;
            }

            int src_index = neighbor_y * cols + neighbor_x;
            int dst_index = y * cols + x;

            float intensity_diff = src[src_index] - src[dst_index];
            float spatial_distance = sqrt((float)(i * i + j * j));
            float weight = gaussianIntensity(intensity_diff, sigma_intensity) * gaussianSpace(spatial_distance, sigma_space);

            intensity_sum += src[src_index] * weight;
            weight_sum += weight;
        }
    }

    dst[y * cols + x] = (unsigned char)(intensity_sum / weight_sum);
}

Mat bilateralFilterCUDA(const Mat& src, int diameter, float sigma_space, float sigma_intensity) {
    Mat filteredImage(src.rows, src.cols, CV_8UC1);
    filteredImage.setTo(0);

    unsigned char* dev_src = nullptr;
    unsigned char* dev_dst = nullptr;
    int rows = src.rows;
    int cols = src.cols;

    cudaMalloc((void**)&dev_src, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_dst, rows * cols * sizeof(unsigned char));

    cudaMemcpy(dev_src, src.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_dst, filteredImage.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);

    dim3 threadsPerBlock(16, 16);
    dim3 numBlocks((cols + threadsPerBlock.x - 1) / threadsPerBlock.x, (rows + threadsPerBlock.y - 1) / threadsPerBlock.y);

    bilateralFilterKernel<<<numBlocks, threadsPerBlock>>>(dev_src, dev_dst, rows, cols, diameter, sigma_space, sigma_intensity);

    cudaMemcpy(filteredImage.data, dev_dst, rows * cols * sizeof(unsigned char), cudaMemcpyDeviceToHost);

    cudaFree(dev_src);
    cudaFree(dev_dst);

    return filteredImage;
}

int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter = atoi(argv[2]);
    float sigma_space = atof(argv[3]);
    float sigma_intensity = atof(argv[4]);
    
    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageCuda = bilateralFilterCUDA(src, diameter, sigma_space, sigma_intensity);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();
    cout<<"Cuda: "<<durationSeq<<"ms \n";
    imwrite("output/cuda_gen.png", filteredImageCuda);
}
```

## Gaussian-spatial precompute in shared memory and Gaussian-intensity precompute  optimization

```python
%%writefile bilateral_cuda_kernel_opt.cu
#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <opencv2/opencv.hpp>
#include <cmath>

using namespace std;
using namespace cv;


// Function to calculate the gaussian weight based on intensity difference
__device__ float gaussianIntensity(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__device__ float gaussianSpace(float spatial_distance, float sigma_space) {
    return expf(-0.5f * (spatial_distance * spatial_distance) / (sigma_space * sigma_space));
}

__global__ void bilateralFilterKernel(const unsigned char* src, unsigned char* dst, int rows, int cols, int diameter, float sigma_space, float sigma_intensity,float* d_spatial_lookup, float* dev_intensity_lookup) {
    int x = blockIdx.x * blockDim.x + threadIdx.x;
    int y = blockIdx.y * blockDim.y + threadIdx.y;

    __shared__ float spatial_lookup[31*31];
    __shared__ float intensity_lookup[256];

    int id_thread_within_block = threadIdx.x*blockDim.y  + threadIdx.y;
    int count = ceil((31*31)/(blockDim.x*blockDim.y));
    int total = 31*31;
    if(id_thread_within_block == blockDim.x*blockDim.y - 1){
      for(int i=id_thread_within_block*count; i<total; i++){
        spatial_lookup[i] = d_spatial_lookup[i];
      }
    }
    else{
      for(int i=0; i<count; i++){
        spatial_lookup[id_thread_within_block*count + i] = d_spatial_lookup[id_thread_within_block*count + i];
      }
    }

    if(id_thread_within_block < 256){
      intensity_lookup[id_thread_within_block] = dev_intensity_lookup[id_thread_within_block];
    }

    if (x >= cols || y >= rows) {
        return;
    }

    __syncthreads();

    float intensity_sum = 0;
    float weight_sum = 0;

    for (int j = -diameter; j <= diameter; j++) {
        for (int i = -diameter; i <= diameter; i++) {
            int neighbor_x = x + i;
            int neighbor_y = y + j;

            if (neighbor_x < 0 || neighbor_y < 0 || neighbor_x >= cols || neighbor_y >= rows) {
                continue;
            }

            int src_index = neighbor_y * cols + neighbor_x;
            int dst_index = y * cols + x;

            int intensity_diff = abs(src[src_index] - src[dst_index]);
            float spatial_weight = spatial_lookup[(i+diameter)*diameter+(j+diameter)];
            float weight = intensity_lookup[intensity_diff] * spatial_weight;
            //float weight = gaussianIntensity(intensity_diff, sigma_intensity) * spatial_weight;


            intensity_sum += src[src_index] * weight;
            weight_sum += weight;
        }
    }

    dst[y * cols + x] = (unsigned char)(intensity_sum / weight_sum);
}

float* computeSpatialLookup(int diameter, float sigma_space) {
    int max_distance = 2*diameter+1;
    float* spatial_lookup = new float[max_distance*max_distance];
    float spatial_distance;
    int t_i, t_j;

    for (int i = 0; i < max_distance; i++) {
      for (int j=0; j< max_distance; j++) {
        t_i = i-diameter;
        t_j = j-diameter;
        spatial_distance = sqrt((float)(t_i * t_i + t_j * t_j));
        spatial_lookup[i*max_distance + j] = expf(-0.5f * spatial_distance / (sigma_space * sigma_space));
      }
    }

    return spatial_lookup;
}

float* computeIntensityLookup(float sigma_intensity) {
  
    float* intensity_lookup = new float[256];
    for (int i = 0; i < 256; i++) {
        intensity_lookup[i] = expf(-0.5f * i*i / (sigma_intensity * sigma_intensity));
    }

    return intensity_lookup;
}

Mat bilateralFilterCUDA(const Mat& src, int diameter, float sigma_space, float sigma_intensity) {
    Mat filteredImage(src.rows, src.cols, CV_8UC1);
    filteredImage.setTo(0);

    unsigned char* dev_src = nullptr;
    unsigned char* dev_dst = nullptr;
    float *dev_spatial_lookup, *dev_intensity_lookup;
    int rows = src.rows;
    int cols = src.cols;
    float* SpatialLookup=computeSpatialLookup(diameter,sigma_space);
    float* IntensityLookup=computeIntensityLookup(sigma_intensity);

    int lookup_row = 2*diameter+1;

    cudaMalloc((void**)&dev_src, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_dst, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_spatial_lookup, sizeof(float)*lookup_row*lookup_row);
    cudaMalloc((void**)&dev_intensity_lookup, sizeof(float)*256);

    cudaMemcpy(dev_src, src.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_dst, filteredImage.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_spatial_lookup, SpatialLookup, sizeof(float)*lookup_row*lookup_row, cudaMemcpyHostToDevice);
    cudaMemcpy(dev_intensity_lookup, IntensityLookup, sizeof(float)*256, cudaMemcpyHostToDevice);

    dim3 threadsPerBlock(16, 16);
    dim3 numBlocks((cols + threadsPerBlock.x - 1) / threadsPerBlock.x, (rows + threadsPerBlock.y - 1) / threadsPerBlock.y);


    bilateralFilterKernel<<<numBlocks, threadsPerBlock>>>(dev_src, dev_dst, rows, cols, diameter, sigma_space, sigma_intensity,dev_spatial_lookup, dev_intensity_lookup);

    cudaMemcpy(filteredImage.data, dev_dst, rows * cols * sizeof(unsigned char), cudaMemcpyDeviceToHost);

    cudaFree(dev_src);
    cudaFree(dev_dst);

    return filteredImage;
}

int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter = atoi(argv[2]);
    float sigma_space = atof(argv[3]);
    float sigma_intensity = atof(argv[4]);
    
    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageCuda = bilateralFilterCUDA(src, diameter, sigma_space, sigma_intensity);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();
    cout<<"Cuda: "<<durationSeq<<"ms \n";
    imwrite("output/cuda_gen_opt2.png", filteredImageCuda);
}
```

## Shared Memory Optimization


```python
%%writefile bilateral_cuda_sm.cu
#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <opencv2/opencv.hpp>

using namespace std;
using namespace cv;

#define THREADS_PER_BLOCK_Y 16
#define THREADS_PER_BLOCK_X 16
#define DIAMETER 30

__constant__ float intensity_weights[256];

// Function to calculate the gaussian weight based on intensity difference
__device__ float gaussianIntensity(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__device__ float gaussianSpace(float spatial_distance, float sigma_space) {
    return expf(-0.5f * (spatial_distance * spatial_distance) / (sigma_space * sigma_space));
}

// Function to calculate the gaussian weight based on spatial distance
__global__ void bilateralFilterKernel(const unsigned char* src, unsigned char* dst, int rows, int cols, int diameter, float sigma_space, float sigma_intensity) {
    int x = blockIdx.x * blockDim.x + threadIdx.x;
    int y = blockIdx.y * blockDim.y + threadIdx.y;

    if (x >= cols || y >= rows) {
        return;
    }

    // Allocate shared memory for a subset of the source image
    __shared__ unsigned char shared_src[THREADS_PER_BLOCK_Y + 2 * DIAMETER][THREADS_PER_BLOCK_X + 2 * DIAMETER];

    // Compute the indices for the shared memory subset
    int shared_x = threadIdx.x + diameter;
    int shared_y = threadIdx.y + diameter;

    // Load the subset of the source image into shared memory
    shared_src[shared_y][shared_x] = src[y * cols + x];
    if (threadIdx.x < diameter) {
        shared_src[shared_y][shared_x - diameter] = (x - diameter >= 0) ? src[y * cols + x - diameter] : 0;
        shared_src[shared_y][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols) ? src[y * cols + x + THREADS_PER_BLOCK_X] : 0;
    }
    if (threadIdx.y < diameter) {
        shared_src[shared_y - diameter][shared_x] = (y - diameter >= 0) ? src[(y - diameter) * cols + x] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x] = (y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x] : 0;
    }
    if (threadIdx.x < diameter && threadIdx.y < diameter) {
        shared_src[shared_y - diameter][shared_x - diameter] = (x - diameter >= 0 && y - diameter >= 0) ? src[(y - diameter) * cols + x - diameter] : 0;
        shared_src[shared_y - diameter][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols && y - diameter >= 0) ? src[(y - diameter) * cols + x + THREADS_PER_BLOCK_X] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x - diameter] = (x - diameter >= 0 && y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x - diameter] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols && y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x + THREADS_PER_BLOCK_X] : 0;
    }

    __syncthreads();

    float intensity_sum = 0;
    float weight_sum = 0;

    for (int j = -diameter; j <= diameter; j++) {
        for (int i = -diameter; i <= diameter; i++){
        int neighbor_shared_x = shared_x + i;
        int neighbor_shared_y = shared_y + j;

        int src_index = (y + j) * cols + (x + i);
        int dst_index = y * cols + x;

        
        
        float spatial_weight = gaussianSpace(sqrt((float)(i * i + j * j)), sigma_space);
        float intensity_diff = shared_src[neighbor_shared_y][neighbor_shared_x] - shared_src[shared_y][shared_x];
        float intensity_weight=gaussianIntensity(intensity_diff,sigma_intensity);

        
    
        float weight =  intensity_weight * spatial_weight;
 

        intensity_sum += shared_src[neighbor_shared_y][neighbor_shared_x] * weight;
        
        weight_sum += weight;
    }
    }
    dst[y * cols + x] = (unsigned char)(intensity_sum / weight_sum);
}


Mat bilateralFilterCUDA(const Mat& src, int diameter, float sigma_space, float sigma_intensity) {
    Mat filteredImage(src.rows, src.cols, CV_8UC1);
    filteredImage.setTo(0);

    unsigned char* dev_src = nullptr;
    unsigned char* dev_dst = nullptr;

    int rows = src.rows;
    int cols = src.cols;


    int lookup_row = 2*diameter+1;

    cudaMalloc((void**)&dev_src, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_dst, rows * cols * sizeof(unsigned char));
    
    cudaMemcpy(dev_src, src.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_dst, filteredImage.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    
    dim3 threadsPerBlock(16, 16);
    dim3 numBlocks((cols + threadsPerBlock.x - 1) / threadsPerBlock.x, (rows + threadsPerBlock.y - 1) / threadsPerBlock.y);


    bilateralFilterKernel<<<numBlocks, threadsPerBlock>>>(dev_src, dev_dst, rows, cols, diameter, sigma_space, sigma_intensity);

    cudaMemcpy(filteredImage.data, dev_dst, rows * cols * sizeof(unsigned char), cudaMemcpyDeviceToHost);

    cudaFree(dev_src);
    cudaFree(dev_dst);

    return filteredImage;
}

int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter = atoi(argv[2]);
    float sigma_space = atof(argv[3]);
    float sigma_intensity = atof(argv[4]);
    
    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageCuda = bilateralFilterCUDA(src, diameter, sigma_space, sigma_intensity);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();
    cout<<"Cuda: "<<durationSeq<<"ms \n";
    imwrite("output/cuda_gen_sm.png", filteredImageCuda);
}
```

## Shared Memory Optimization + kernel

```python
%%writefile bilateral_cuda_sm_kernel.cu
#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <opencv2/opencv.hpp>

using namespace std;
using namespace cv;

#define THREADS_PER_BLOCK_Y 16
#define THREADS_PER_BLOCK_X 16
#define DIAMETER 30

// Function to calculate the gaussian weight based on intensity difference
__device__ float gaussianIntensity(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__device__ float gaussianSpace(float spatial_distance, float sigma_space) {
    return expf(-0.5f * (spatial_distance * spatial_distance) / (sigma_space * sigma_space));
}

float gaussianIntensity_cpu(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__global__ void bilateralFilterKernel(const unsigned char* src, unsigned char* dst, int rows, int cols, int diameter, float sigma_space, float sigma_intensity, float* spatial_lookup,float* IntensityLookup) {
    int x = blockIdx.x * blockDim.x + threadIdx.x;
    int y = blockIdx.y * blockDim.y + threadIdx.y;

    if (x >= cols || y >= rows) {
        return;
    }

    // Allocate shared memory for a subset of the source image
    __shared__ unsigned char shared_src[THREADS_PER_BLOCK_Y + 2 * DIAMETER][THREADS_PER_BLOCK_X + 2 * DIAMETER];

    // Compute the indices for the shared memory subset
    int shared_x = threadIdx.x + diameter;
    int shared_y = threadIdx.y + diameter;

    // Load the subset of the source image into shared memory
    shared_src[shared_y][shared_x] = src[y * cols + x];
    if (threadIdx.x < diameter) {
        shared_src[shared_y][shared_x - diameter] = (x - diameter >= 0) ? src[y * cols + x - diameter] : 0;
        shared_src[shared_y][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols) ? src[y * cols + x + THREADS_PER_BLOCK_X] : 0;
    }
    if (threadIdx.y < diameter) {
        shared_src[shared_y - diameter][shared_x] = (y - diameter >= 0) ? src[(y - diameter) * cols + x] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x] = (y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x] : 0;
    }
    if (threadIdx.x < diameter && threadIdx.y < diameter) {
        shared_src[shared_y - diameter][shared_x - diameter] = (x - diameter >= 0 && y - diameter >= 0) ? src[(y - diameter) * cols + x - diameter] : 0;
        shared_src[shared_y - diameter][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols && y - diameter >= 0) ? src[(y - diameter) * cols + x + THREADS_PER_BLOCK_X] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x - diameter] = (x - diameter >= 0 && y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x - diameter] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols && y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x + THREADS_PER_BLOCK_X] : 0;
    }

    int id_thread_within_block = threadIdx.x*blockDim.y  + threadIdx.y;

    __syncthreads();

    float intensity_sum = 0;
    float weight_sum = 0;

    for (int j = -diameter; j <= diameter; j++) {
        for (int i = -diameter; i <= diameter; i++){
        int neighbor_shared_x = shared_x + i;
        int neighbor_shared_y = shared_y + j;

        int src_index = (y + j) * cols + (x + i);
        int dst_index = y * cols + x;

        int intensity_diff = abs(shared_src[neighbor_shared_y][neighbor_shared_x] - shared_src[shared_y][shared_x]);
        
        //float spatial_weight = gaussianSpace(sqrt((float)(i * i + j * j)), sigma_space);
        float spatial_weight = spatial_lookup[(i + diameter) * (2*diameter+1) + (j + diameter)];
        
        int intensity_diff_int=intensity_diff;
        //intensity_diff_int= (intensity_diff_int<0)? (-intensity_diff_int): intensity_diff_int;
        float intensity_weight=IntensityLookup[intensity_diff_int];
        //float intensity_weight=gaussianIntensity(intensity_diff,sigma_intensity);

        
    
        float weight =  intensity_weight * spatial_weight;
 

        intensity_sum += shared_src[neighbor_shared_y][neighbor_shared_x] * weight;
        
        weight_sum += weight;
    }
    }
    dst[y * cols + x] = (unsigned char)(intensity_sum / weight_sum);
}


float* computeSpatialLookup(int diameter, float sigma_space) {
    int max_distance = 2*diameter+1;
    float* spatial_lookup = new float[max_distance*max_distance];
    float spatial_distance;
    int t_i, t_j;

    for (int i = 0; i < max_distance; i++) {
      for (int j=0; j<max_distance; j++) {
        t_i = i-diameter;
        t_j = j-diameter;
        spatial_distance = sqrt((float)(t_i * t_i + t_j * t_j));
        spatial_lookup[i*max_distance + j] = expf(-0.5f *spatial_distance* spatial_distance / (sigma_space * sigma_space));
      }
    }

    return spatial_lookup;
}
float* computeIntensityLookup(float sigma_intensity) {
  
    float* intensity_lookup = new float[256];
    for (int i = 0; i < 256; i++) {
        intensity_lookup[i] = expf(-0.5f * i*i/ (sigma_intensity * sigma_intensity));
    }

    return intensity_lookup;
}

Mat bilateralFilterCUDA(const Mat& src, int diameter, float sigma_space, float sigma_intensity) {
    Mat filteredImage(src.rows, src.cols, CV_8UC1);
    filteredImage.setTo(0);

    unsigned char* dev_src = nullptr;
    unsigned char* dev_dst = nullptr;
    float* dev_spatial_lookup, *dev_intensity_lookup;
    int rows = src.rows;
    int cols = src.cols;
    float* SpatialLookup=computeSpatialLookup(diameter,sigma_space);
    float* IntensityLookup=computeIntensityLookup(sigma_intensity);


    int lookup_row = 2*diameter+1;

    cudaMalloc((void**)&dev_src, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_dst, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_spatial_lookup, sizeof(float)*lookup_row*lookup_row);
    cudaMalloc((void**)&dev_intensity_lookup, sizeof(float)*256);

    cudaMemcpy(dev_src, src.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_dst, filteredImage.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_spatial_lookup, SpatialLookup, sizeof(float)*lookup_row*lookup_row, cudaMemcpyHostToDevice);
    cudaMemcpy(dev_intensity_lookup, IntensityLookup, sizeof(float)*256, cudaMemcpyHostToDevice);

    dim3 threadsPerBlock(16, 16);
    dim3 numBlocks((cols + threadsPerBlock.x - 1) / threadsPerBlock.x, (rows + threadsPerBlock.y - 1) / threadsPerBlock.y);


    bilateralFilterKernel<<<numBlocks, threadsPerBlock>>>(dev_src, dev_dst, rows, cols, diameter, sigma_space, sigma_intensity,dev_spatial_lookup,dev_intensity_lookup);

    cudaMemcpy(filteredImage.data, dev_dst, rows * cols * sizeof(unsigned char), cudaMemcpyDeviceToHost);

    cudaFree(dev_src);
    cudaFree(dev_dst);

    return filteredImage;
}

int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter = atoi(argv[2]);
    float sigma_space = atof(argv[3]);
    float sigma_intensity = atof(argv[4]);
    
    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageCuda = bilateralFilterCUDA(src, diameter, sigma_space, sigma_intensity);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();
    cout<<"Cuda: "<<durationSeq<<"ms \n";
    imwrite("output/cuda_gen_sm2.png", filteredImageCuda);
}
```

## Sequential and Parallel

```python
%%writefile nlmeans_sequential.cpp
#include <iostream>
#include <stdio.h>
#include <opencv2/opencv.hpp>
#include <thread>
#include <chrono>

using namespace std;
using namespace cv;

// Function to calculate the gaussian weight based on spatial distance
float gaussian(float x, float sigma) {
    return expf(-0.5f * (x*x) / (sigma * sigma));
}

void nonLocalBilateralFilterKernel(int x, int y, Mat src, Mat dst, int rows, int cols, int diameter, float sigma_space, float sigma_intensity, float sigma_range) {
    if (x >= cols || y >= rows) {
        return;
    }
    float intensity_sum = 0;
    float weight_sum = 1;

    int half_diameter = diameter / 2;
    int patch_size = diameter * diameter;

    for (int j = 0; j < rows; j++) {
        for (int i = 0; i < cols; i++) {
            int neighbor_x = i;
            int neighbor_y = j;
            float spatial_distance = sqrtf((float)((x - neighbor_x) * (x - neighbor_x) + (y - neighbor_y) * (y - neighbor_y)));

            float range_weight = 1.0;
            if (spatial_distance < half_diameter) {
                float ssd = 0.0;
                for (int r = -half_diameter; r <= half_diameter; r++) {
                    for (int c = -half_diameter; c <= half_diameter; c++) {
                        int patch_x = x + c;
                        int patch_y = y + r;
                        if (patch_x >= 0 && patch_x < cols && patch_y >= 0 && patch_y < rows) {
                            int patch_src_index = patch_y * cols + patch_x;
                            int patch_dst_index = y * cols + x;
                            float patch_intensity_diff = src.at<uchar>(patch_x, patch_y) - src.at<uchar>(x, y);
                            float patch_range_diff = dst.at<uchar>(patch_x, patch_y) - dst.at<uchar>(x, y);
                            ssd += patch_intensity_diff * patch_intensity_diff + patch_range_diff * patch_range_diff;
                        }
                    }
                }
                range_weight = expf(ssd / (sigma_range * sigma_range * patch_size));
            }
            int intensity_diff = src.at<uchar>(neighbor_x, neighbor_y) - src.at<uchar>(x, y);
            float spatial_weight = expf(-0.5f * (spatial_distance*spatial_distance) / (sigma_space * sigma_space));
            //float spatial_weight = spatial_lookup[(i + diameter) * (2*diameter+1) + (j + diameter)];
            float intensity_weight=expf(-0.5f * (intensity_diff*intensity_diff) / (sigma_intensity * sigma_intensity));
            //float intensity_weight=IntensityLookup[abs(intensity_diff)];
            float weight = intensity_weight*range_weight;
            
            intensity_sum += src.at<uchar>(neighbor_x, neighbor_y) * weight;
            weight_sum += weight; 
        }
    }
    dst.at<double>(x, y) = (unsigned char)(intensity_sum / weight_sum);
}
float* computeSpatialLookup(int diameter, float sigma_space) {
    int max_distance = 2*diameter+1;
    float* spatial_lookup = new float[max_distance*max_distance];
    float spatial_distance;
    int t_i, t_j;

    for (int i = 0; i < max_distance; i++) {
      for (int j=0; j< max_distance; j++) {
        t_i = i-diameter;
        t_j = j-diameter;
        spatial_distance = sqrt((float)(t_i * t_i + t_j * t_j));
        spatial_lookup[i*max_distance + j] = expf(-0.5f * spatial_distance*spatial_distance / (sigma_space * sigma_space));
      }
    }

    return spatial_lookup;
}
float* computeIntensityLookup(float sigma_intensity) {
  
    float* intensity_lookup = new float[256];
    for (int i = 0; i < 256; i++) {
        intensity_lookup[i] = expf(-0.5f * i*i/ (sigma_intensity * sigma_intensity));
    }

    return intensity_lookup;
}

Mat nonLocalBilateralFilterParallel(Mat source, int diameter, double sigmaI, double sigmaS,int numThreads, double sigmaR) {
    Mat filteredImage = Mat::zeros(source.rows, source.cols, CV_64F);
    int radius = 0;

    int width = source.cols;
    int height = source.rows;

    //int numThreads = thread::hardware_concurrency();
    
    vector<thread> threads;

    int rowsPerThread = height / numThreads;

    int start = radius;
    int end = start + rowsPerThread;

    for (int t = 0; t < numThreads; t++) {
        if (t == numThreads - 1) {
            end = height - radius;
        }
        
        threads.push_back(thread([&, start, end]() {
            for (int i = start; i < end; i++) {
                for (int j = radius; j < width - radius; j++) {
                    nonLocalBilateralFilterKernel(i, j, source, filteredImage, height, width, diameter, sigmaI, sigmaS, sigmaR);
                }
            }
        }));

        start = end;
        end = start + rowsPerThread;
    }

    for (auto &t : threads) {
        t.join();
    }

    return filteredImage;
}

Mat nonLocalBilateralFilterSeq(Mat src, int diameter, float sigma_space, float sigma_intensity, float sigma_range) {
    Mat filteredImage(src.rows, src.cols, CV_8UC1);
    filteredImage.setTo(0);
    int rows = src.rows;
    int cols = src.cols;


    int lookup_row = 2*diameter+1;

    for(int x=0; x<rows; x++){
      for(int y=0; y<cols; y++){
          nonLocalBilateralFilterKernel(x, y, src, filteredImage, rows, cols, diameter, sigma_space, sigma_intensity, sigma_range);
      }
    }

    return filteredImage;
}

int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter = atoi(argv[2]);
    float sigma_space = atof(argv[3]);
    float sigma_intensity = atof(argv[4]);
    float sigma_range=atof(argv[4]);
    int numThreads = 10;

    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }
    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageSeq = nonLocalBilateralFilterSeq(src, diameter, sigma_space, sigma_intensity, sigma_range);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();

    start = chrono::high_resolution_clock::now();
    Mat filteredImageParallel = nonLocalBilateralFilterParallel(src, diameter, sigma_intensity,sigma_space,numThreads, sigma_range);
    end = chrono::high_resolution_clock::now();
    auto durationParallel = chrono::duration_cast<chrono::milliseconds>(end - start).count();

    cout<<"Sequential: "<<durationSeq<<"ms \n";
    cout<<"Parallel: "<<durationParallel<<"ms"<<" using "<<numThreads<<" threads\n";

    imwrite("output/nl_means_seq.png", filteredImageSeq);
    imwrite("output/nl_means_parallel.png", filteredImageParallel);
}


```

## Sequential and CPU Parallel implementations

```python
%%writefile bilateral_adaptive_sequential.cpp

#include <iostream>
#include <stdio.h>
#include <opencv2/opencv.hpp>
#include <thread>
#include <chrono>

using namespace cv;
using namespace std;

float distance(int x, int y, int i, int j) {
    return float(sqrt(pow(x - i, 2) + pow(y - j, 2)));
}

float gaussian(float x, double sigma) {
    //return exp(-(pow(x, 2))/(2 * pow(sigma, 2))) / (2 * CV_PI * pow(sigma, 2));
    return expf(-0.5f * (x*x) / (sigma* sigma));

}
void applyAdaptiveFilter(Mat source, Mat filteredImage, int x, int y, int diameter, double sigmaI, double sigmaS) {
    double iFiltered = 0;
    double wP = 0;
    int neighbor_x = 0;
    int neighbor_y = 0;
    int half = diameter / 2;

    int rows = source.rows;
    int cols = source.cols;

    float variance_sum = 0, variance_weight_sum = 0;
    float src_center = source.at<uchar>(x, y);

    int intensity_sum = 0, weight_sum = 0;

    for (int i = 0; i <= diameter; i++) {
        for (int j = 0; j <= diameter; j++) {
            neighbor_x = x - (half - i);
            neighbor_y = y - (half - j);

            float src_neighbor = source.at<uchar>(neighbor_x, neighbor_y);
            float intensity_diff = src_neighbor - src_center;
            float spatial_distance = sqrt((float)(i * i + j * j));
            
            float weight = gaussian(intensity_diff, sigmaI) * gaussian(spatial_distance, sigmaS);

            // Accumulate intensity and weight sums for filtering
            intensity_sum += src_neighbor * weight;
            weight_sum += weight;

            // Accumulate variance and weight sums for calculating variance
            float intensity_squared_diff = intensity_diff * intensity_diff;
            variance_sum += intensity_squared_diff * weight;
            variance_weight_sum += weight;
        }
    }

    float variance = variance_sum / variance_weight_sum;

    // Calculate new sigma_intensity using variance
    float new_sigma_intensity = sqrt(max(variance + 0.01, 0.0)) / 2;

    intensity_sum = 0;
    weight_sum = 0;

    for(int i = 0; i < diameter; i++) {
        for(int j = 0; j < diameter; j++) {
            neighbor_x = x - (half - i);
            neighbor_y = y - (half - j);
            double gi = gaussian(source.at<uchar>(neighbor_x, neighbor_y) - source.at<uchar>(x, y), new_sigma_intensity);
            float spatial_distance = sqrt((float)(i * i + j * j));
            double gs = gaussian(spatial_distance, sigmaS);
            double w = gi * gs;
            iFiltered = iFiltered + source.at<uchar>(neighbor_x, neighbor_y) * w;
            wP = wP + w;
        }
    }
    iFiltered = iFiltered / wP;
    filteredImage.at<double>(x, y) = (unsigned char)iFiltered;
}


Mat bilateralAdaptiveSeq(Mat source, int diameter, double sigmaI, double sigmaS) {
    Mat filteredImage = Mat::zeros(source.rows,source.cols,CV_64F);
    int radius=0;

    int width = source.cols;
    int height = source.rows;

    for(int i = radius; i < height - radius; i++) {
        for(int j = radius; j < width - radius; j++) {
            applyAdaptiveFilter(source, filteredImage, i, j, diameter, sigmaI, sigmaS);
        }
    }
    return filteredImage;
}


Mat bilateralAdaptiveParallel(Mat source, int diameter, double sigmaI, double sigmaS,int numThreads) {
    Mat filteredImage = Mat::zeros(source.rows, source.cols, CV_64F);
    int radius = 0;

    int width = source.cols;
    int height = source.rows;

    //int numThreads = thread::hardware_concurrency();
    
    vector<thread> threads;

    int rowsPerThread = height / numThreads;

    int start = radius;
    int end = start + rowsPerThread;

    for (int t = 0; t < numThreads; t++) {
        if (t == numThreads - 1) {
            end = height - radius;
        }

        threads.push_back(thread([&, start, end]() {
            for (int i = start; i < end; i++) {
                for (int j = radius; j < width - radius; j++) {
                    applyAdaptiveFilter(source, filteredImage, i, j, diameter, sigmaI, sigmaS);
                }
            }
        }));

        start = end;
        end = start + rowsPerThread;
    }

    for (auto &t : threads) {
        t.join();
    }

    return filteredImage;
}


int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter= atoi(argv[2]);
    double sigma_space= atoi(argv[3]);
    double sigma_intensity= atoi(argv[4]);
    int numThreads = 10;
    

    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageSeq = bilateralAdaptiveSeq(src, diameter, sigma_intensity,sigma_space);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();

    start = chrono::high_resolution_clock::now();
    Mat filteredImageParallel = bilateralAdaptiveParallel(src, diameter, sigma_intensity,sigma_space,numThreads);
    end = chrono::high_resolution_clock::now();
    auto durationParallel = chrono::duration_cast<chrono::milliseconds>(end - start).count();

    cout<<"Sequential: "<<durationSeq<<"ms \n";
    cout<<"Parallel: "<<durationParallel<<"ms"<<" using "<<numThreads<<" threads\n";

    imwrite("output/adaptive_seq.png", filteredImageSeq);
    imwrite("output/adaptive_parallel.png", filteredImageParallel);

    return 0;
}
```

## GPU Shared memory (Img) and caching kernels

```python
%%writefile bilateral_adaptive_opt.cu
#include <iostream>
#include <chrono>
#include <cuda_runtime.h>
#include <opencv2/opencv.hpp>

using namespace std;
using namespace cv;

#define THREADS_PER_BLOCK_Y 16
#define THREADS_PER_BLOCK_X 16
#define DIAMETER 30

__constant__ float intensity_weights[256];

// Function to calculate the gaussian weight based on intensity difference
__device__ float gaussianIntensity(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__device__ float gaussianSpace(float spatial_distance, float sigma_space) {
    return expf(-0.5f * (spatial_distance * spatial_distance) / (sigma_space * sigma_space));
}

float gaussianIntensity_cpu(float intensity_diff, float sigma_intensity) {
    return expf(-0.5f * (intensity_diff * intensity_diff) / (sigma_intensity * sigma_intensity));
}

// Function to calculate the gaussian weight based on spatial distance
__global__ void bilateralFilterKernel(const unsigned char* src, unsigned char* dst, int rows, int cols, int diameter, float sigma_space, float sigma_intensity, float* spatial_lookup,float* IntensityLookup) {
    int x = blockIdx.x * blockDim.x + threadIdx.x;
    int y = blockIdx.y * blockDim.y + threadIdx.y;

    if (x >= cols || y >= rows) {
        return;
    }

    // Allocate shared memory for a subset of the source image
    __shared__ unsigned char shared_src[THREADS_PER_BLOCK_Y + 2 * DIAMETER][THREADS_PER_BLOCK_X + 2 * DIAMETER];

    // Compute the indices for the shared memory subset
    int shared_x = threadIdx.x + diameter;
    int shared_y = threadIdx.y + diameter;

    // Load the subset of the source image into shared memory
    shared_src[shared_y][shared_x] = src[y * cols + x];
    if (threadIdx.x < diameter) {
        shared_src[shared_y][shared_x - diameter] = (x - diameter >= 0) ? src[y * cols + x - diameter] : 0;
        shared_src[shared_y][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols) ? src[y * cols + x + THREADS_PER_BLOCK_X] : 0;
    }
    if (threadIdx.y < diameter) {
        shared_src[shared_y - diameter][shared_x] = (y - diameter >= 0) ? src[(y - diameter) * cols + x] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x] = (y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x] : 0;
    }
    if (threadIdx.x < diameter && threadIdx.y < diameter) {
        shared_src[shared_y - diameter][shared_x - diameter] = (x - diameter >= 0 && y - diameter >= 0) ? src[(y - diameter) * cols + x - diameter] : 0;
        shared_src[shared_y - diameter][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols && y - diameter >= 0) ? src[(y - diameter) * cols + x + THREADS_PER_BLOCK_X] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x - diameter] = (x - diameter >= 0 && y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x - diameter] : 0;
        shared_src[shared_y + THREADS_PER_BLOCK_Y][shared_x + THREADS_PER_BLOCK_X] = (x + THREADS_PER_BLOCK_X < cols && y + THREADS_PER_BLOCK_Y < rows) ? src[(y + THREADS_PER_BLOCK_Y) * cols + x + THREADS_PER_BLOCK_X] : 0;
    }

    __syncthreads();

    float intensity_sum = 0, weight_sum = 0;
    float src_center = src[y * cols + x];

    // Calculate variance in a window
    float variance_sum = 0, variance_weight_sum = 0;
    int window_size = (2 * diameter + 1) * (2 * diameter + 1);

    for (int j = -diameter; j <= diameter; j++) {
        for (int i = -diameter; i <= diameter; i++) {
            int neighbor_x = shared_x + i;
            int neighbor_y = shared_y + j;

            if (neighbor_x < 0 || neighbor_y < 0 || neighbor_x >= cols || neighbor_y >= rows) continue;

            //float src_neighbor = src[neighbor_y * cols + neighbor_x];
            float src_neighbor=src[y * cols + x];
            //float intensity_diff = src_neighbor - src_center;
            int intensity_diff = (int)abs(shared_src[neighbor_y][neighbor_x] - shared_src[shared_y][shared_x]);
            float spatial_distance = sqrt((float)(i * i + j * j));
            float spatial_weight = spatial_lookup[(i + diameter) * (2*diameter+1) + (j + diameter)];
            //float weight = gaussianIntensity(intensity_diff, sigma_intensity) * gaussianSpace(spatial_distance, sigma_space);
            float weight =IntensityLookup[intensity_diff] * spatial_weight;

            // Accumulate intensity and weight sums for filtering
            intensity_sum += src_neighbor * weight;
            weight_sum += weight;

            // Accumulate variance and weight sums for calculating variance
            float intensity_squared_diff = intensity_diff * intensity_diff;
            variance_sum += intensity_squared_diff * weight;
            variance_weight_sum += weight;
        }
    }

    // Calculate variance in the window
    float variance = variance_sum / variance_weight_sum;

    // Calculate new sigma_intensity using variance
    float new_sigma_intensity = sqrt(max(variance + 0.01, 0.f)) / 2;

    intensity_sum = 0;
    weight_sum = 0;

    for (int j = -diameter; j <= diameter; j++) {
        for (int i = -diameter; i <= diameter; i++){
        int neighbor_shared_x = shared_x + i;
        int neighbor_shared_y = shared_y + j;

        int src_index = (y + j) * cols + (x + i);
        int dst_index = y * cols + x;

        int intensity_diff = (int)abs(shared_src[neighbor_shared_y][neighbor_shared_x] - shared_src[shared_y][shared_x]);
        float spatial_weight = spatial_lookup[(i + diameter) * (2*diameter+1) + (j + diameter)];
        float intensity_weight=gaussianIntensity(intensity_diff, new_sigma_intensity);
        //float spatial_distance = sqrt((float)(i * i + j * j));
        //float weight = gaussianIntensity(intensity_diff, new_sigma_intensity) * gaussianSpace(spatial_distance, sigma_space);
        float weight =  intensity_weight * spatial_weight;
 

        intensity_sum += shared_src[neighbor_shared_y][neighbor_shared_x] * weight;
        
        weight_sum += weight;
    }
    }
    dst[y * cols + x] = (unsigned char)(intensity_sum / weight_sum);
}


float* computeSpatialLookup(int diameter, float sigma_space) {
    int max_distance = 2*diameter+1;
    float* spatial_lookup = new float[max_distance*max_distance];
    float spatial_distance;
    int t_i, t_j;

    for (int i = 0; i < max_distance; i++) {
      for (int j=0; j<max_distance; j++) {
        t_i = i-diameter;
        t_j = j-diameter;
        spatial_distance = sqrt((float)(t_i * t_i + t_j * t_j));
        spatial_lookup[i*max_distance + j] = expf(-0.5f * spatial_distance* spatial_distance / (sigma_space * sigma_space));
      }
    }

    return spatial_lookup;
}
float* computeIntensityLookup(float sigma_intensity) {
  
    float* intensity_lookup = new float[256];
    for (int i = 0; i < 256; i++) {
        intensity_lookup[i] = expf(-0.5f * i*i/ (sigma_intensity * sigma_intensity));
    }

    return intensity_lookup;
}


Mat bilateralFilterCUDA(const Mat& src, int diameter, float sigma_space, float sigma_intensity) {
    Mat filteredImage(src.rows, src.cols, CV_8UC1);
    filteredImage.setTo(0);

    unsigned char* dev_src = nullptr;
    unsigned char* dev_dst = nullptr;
    float* dev_spatial_lookup,*dev_intensity_lookup;
    int rows = src.rows;
    int cols = src.cols;
    float* SpatialLookup=computeSpatialLookup(diameter,sigma_space);
    float* IntensityLookup=computeIntensityLookup(sigma_intensity);


    int lookup_row = 2*diameter+1;

    cudaMalloc((void**)&dev_src, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_dst, rows * cols * sizeof(unsigned char));
    cudaMalloc((void**)&dev_spatial_lookup, sizeof(float)*lookup_row*lookup_row);
    cudaMalloc((void**)&dev_intensity_lookup, sizeof(float)*256);

    cudaMemcpy(dev_src, src.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_dst, filteredImage.data, rows * cols * sizeof(unsigned char), cudaMemcpyHostToDevice);
    cudaMemcpy(dev_spatial_lookup, SpatialLookup, sizeof(float)*lookup_row*lookup_row, cudaMemcpyHostToDevice);
    cudaMemcpy(dev_intensity_lookup, IntensityLookup, sizeof(float)*256, cudaMemcpyHostToDevice);

    dim3 threadsPerBlock(16, 16);
    dim3 numBlocks((cols + threadsPerBlock.x - 1) / threadsPerBlock.x, (rows + threadsPerBlock.y - 1) / threadsPerBlock.y);


    bilateralFilterKernel<<<numBlocks, threadsPerBlock>>>(dev_src, dev_dst, rows, cols, diameter, sigma_space, sigma_intensity,dev_spatial_lookup,dev_intensity_lookup);

    cudaMemcpy(filteredImage.data, dev_dst, rows * cols * sizeof(unsigned char), cudaMemcpyDeviceToHost);

    cudaFree(dev_src);
    cudaFree(dev_dst);

    return filteredImage;
}

int main(int argc, char** argv ) {
    Mat src;
    src = imread( argv[1], 0 );
    int diameter = atoi(argv[2]);
    float sigma_space = atof(argv[3]);
    float sigma_intensity = atof(argv[4]);
    
    if ( !src.data )
    {
        printf("No image data \n");
        return -1;
    }

    auto start = chrono::high_resolution_clock::now();
    Mat filteredImageCuda = bilateralFilterCUDA(src, diameter, sigma_space, sigma_intensity);
    auto end = chrono::high_resolution_clock::now();
    auto durationSeq = chrono::duration_cast<chrono::milliseconds>(end - start).count();
    cout<<"Cuda: "<<durationSeq<<"ms \n";
    imwrite("output/cuda_gen_adaptive_opt.png", filteredImageCuda);
}
```
